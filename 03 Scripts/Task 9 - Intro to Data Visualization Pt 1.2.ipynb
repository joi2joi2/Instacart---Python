#### 5. If-Statement using For-Loops to create 'new_customer' column in the 'orders_products_dervived_2' dataframe

```python
# String for Main Project Folder
path = r'C:\Users\ifari\12-2022 Instacart Basket Analysis'
```

```python
# String to import Data Folder containing orders_products_derived.pkl
df_ords_prods_d = pd.read_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_derived_2.pkl'))
```

```python
# Dimensions of the 'order_produces_merged_2' dataframe
df_ords_prods_d.shape
```

```python
# Check for missing values in the df_ords_prods_d
df_ords_prods_d.isnull().sum()
```

```python
# Defining the 'new_customer' function
result = []

for value in df_ords_prods_d["days_since_prior_order"]:
  if value >= 0:
    result.append("False")
  else:
    result.append("True")
```

```python
# Define the 'new_customer' function
df_ords_prods_d['new_customer'] = result
```

```python
# Frequency of the values in the 'new_customer' column
df_ords_prods_d['new_customer'].value_counts(dropna = False)
```

```python
# Length of the 'new_customer' column
len(result)
```

#### 5. Merge 'customers_checked' and 'orders_products_derived_3'

```python
# String to import Data Folder containing orders_products_derived.pkl
df_cust_checked = pd.read_pickle(os.path.join(path, '02 Data','Prepared Data', 'customers_checked.pkl'))
```

```python
# String to import Data Folder containing orders_products_derived.pkl
df_ords_prods_derived = pd.read_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_derived_3.pkl'))
```

```python
# Checking datatypes of the 'order_products' dataframe
df_ords_prods_derived.info()
```

```python
# Checking datatypes of the 'customer' dataframe
df_cust_checked.info()
```

```python
# Changing the 'user_id' datatype in 'order_products' and 'customer' dataframes
df_ords_prods_derived['user_id'] = df_ords_prods_derived['user_id'].astype('int32')
df_cust_checked['user_id'] = df_cust_checked['user_id'].astype('int32')
```

```python
# Dimensions of 'order_products' dataframe
df_ords_prods_derived.shape
```

```python
# Dimensions of 'customer' dataframe
df_cust_checked.shape
```

```python
# First 5 rows of the combined dataframes of 'order_products' and 'customer'
df_merged = df_ords_prods_derived.merge(df_cust_checked, on = 'user_id', indicator = True)
df_merged.head()
```

```python
# Dimenstions of merged dataframe
df_merged.shape
```

```python
# Frequency of the values in the '_merge' column
df_merged['_merge'].value_counts()
```

```python
# Drop '_merge' column
df_merged.drop(columns = ['_merge'])
```

#### 6. Export Merged Dataframe

```python
# Export data in pickle format
df_merged.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_all.pkl'))
```
